```python
import pandas as pd
import glob

folderName = 'res_gossiping'
# glob.glob(folderName + '/*txt')

data = list()
for path in glob.glob(folderName + '/*txt'):
    with open(path, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('---split---')[1].split('\n')[1:-1]
        data.append(tmpList)
# data
columns = [r.split(': ')[0] for r in data[0]]
# columns

df = pd.DataFrame(data=data, columns=columns)
df
```
